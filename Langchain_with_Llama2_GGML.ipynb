```python
response = llm_chain.run("""Explain
--- \index.css ---
html{
    height: 100%;
}
body{

    display: flex;
    flex-wrap: wrap;
    min-height: 100%;
    margin: 0;
    flex-direction: row;
}
.top-container{
    padding: 0.25rem;
    min-height: 2rem;
    position: absolute;
    min-width: 40%;
    width: calc(100% - 0.5rem);
    display: flex;
    box-shadow: 2px 2px 2px grey;
}
.top-container-left-actions{

}
.search-input-container{
    max-width: calc(60% - 0.5rem);
    flex-grow: 1;
    display: flex;
}
.search-input-container>input{
    flex: 1;
}
.top-container-right-actions{

}
.web-windows-container{
    flex-grow: 1;
    display: flex;
    flex-wrap: wrap;
    margin: 0;
    justify-content: center;
    align-items: center;
    align-content: center;
}

.container{
  flex: 1;
  min-width: calc(50% - 2rem);
  margin: 1rem;
  box-shadow: 1px 2px 6px 0px #377b84;
  max-width: 300px;
  max-height: 240px;
  min-height: 240px;
  display: flex;
  flex-direction: column;
}

.container.active{
  position: absolute;
  height: 100%;
  width: 100%;
  top: 0;
  margin: 0;
  max-width: 100%;
  max-height: 100%;
}

.container .title {
    background: #1abc9c;
    color: white;
    font-family: sans-serif;
    display: flex;
    width: calc(100%);
    align-items: center;
}
.container .title .material-icons{
    padding: 0.25rem 0.25rem;
}

.container .title .actions{
  cursor: pointer;
}
.container.active .title .actions.maximize{
  display: none;
}
.container:not(.active) .title .actions.minimize{
  display: none;
}
.container.new .title .actions{
  display: none;
}

.container .title .title-text{
    flex-grow: 1;
    padding: 0.25rem 0.25rem;
    white-space: nowrap;
    overflow: hidden;
    text-overflow: ellipsis;
}

.container .video ,.container .content{
    flex-grow: 1;
    cursor: pointer;
}



/*Automcplete*/
.autocomplete {
  /*the container must be positioned relative:*/
  position: relative;
  display: inline-block;
}
.autocomplete input {
  border: 1px solid transparent;
  background-color: #f1f1f1;
  padding: 10px;
  font-size: 16px;
}
.autocomplete input[type=text] {
  background-color: #f1f1f1;
  width: 100%;
}
.autocomplete input[type=submit] {
  background-color: DodgerBlue;
  color: #fff;
}
.autocomplete-items {
  position: absolute;
  border: 1px solid #d4d4d4;
  border-bottom: none;
  border-top: none;
  z-index: 99;
  /*position the autocomplete items to be the same width as the container:*/
  top: 100%;
  left: 0;
  right: 0;
}
.autocomplete-items div {
  padding: 10px;
  cursor: pointer;
  background-color: #fff;
  border-bottom: 1px solid #d4d4d4;
}
.autocomplete-items div:hover {
  /*when hovering an item:*/
  background-color: #e9e9e9;
}
.autocomplete-active {
  /*when navigating through the items using the arrow keys:*/
  background-color: DodgerBlue !important;
  color: #ffffff;
}

--- \index.html ---
<!DOCTYPE html>
<html lang="en">
<head>
  <title>OttO</title>
  <link rel="stylesheet" href="index.css"/>
  <link rel="stylesheet" href="node_modules/material-icons/iconfont/material-icons.css"/>
</head>
<body>
</body>
</html>

--- \indexleft.html ---
<!DOCTYPE html>
<html lang="en">

<head>
    <title>Document</title>
    <link rel="stylesheet" href="index.css" />
    <style>
        body {
            background: grey;
        }
    </style>
</head>

<body>
    Hello Left
</body>

</html>

--- \indexright.html ---
<!DOCTYPE html>
<html lang="en">

<head>
    <title>Document</title>
    <link rel="stylesheet" href="index.css" />
    <style>
        body {
            background: yellow;
        }
    </style>
</head>

<body>
    Hello Right
</body>

</html>

--- \inject.js ---
console.log('Executed from main')
window.addEventListener('DOMContentLoaded', (event) => {
    window.openDevTools()
    console.log('DOM fully loaded and parsed');
    if (document.body && document.body.innerHTML)
        document.body.innerHTML = document.body.innerHTML.replaceAll("Log", "LLoogg");
    document.body.setAttribute('style',"")
    // index.js
    const clickedElements = document.querySelectorAll("*");
    let getElementSelector = (el) => {
        if (!(el instanceof Element))
            return;
        var path = [];
        while (el.nodeType === Node.ELEMENT_NODE) {
            var selector = el.nodeName.toLowerCase();
            if (el.id) {
                selector += '#' + el.id;
                path.unshift(selector);
                break;
            } else {
                var sib = el, nth = 1;
                while (sib = sib.previousElementSibling) {
                    if (sib.nodeName.toLowerCase() == selector)
                        nth++;
                }
                if (nth != 1)
                    selector += ":nth-of-type(" + nth + ")";
            }
            path.unshift(selector);
            el = el.parentNode;
        }
        return path.join(" > ");
    }
    [...clickedElements].forEach(n => {

        n.addEventListener("click", (event) => {
            if (event.currentTarget == event.target)
                window.theNameYouWant.send("element-click", {
                    type: event.type,
                    selector: getElementSelector(event.currentTarget),
                    currentTarget: event.currentTarget.tagName,
                    target: event.target.tagName
                })
        });

        n.addEventListener("change", (event) => {
            if (event.currentTarget == event.target)
                window.theNameYouWant.send("element-change", {
                    type: event.type,
                    selector: getElementSelector(event.currentTarget),
                    currentTarget: event.currentTarget.tagName,
                    target: event.target.tagName
                })
        });

    })

});
        // Empty

--- \lib\autoscripts\autoscript-preload.js ---
// In the preload script.
const { contextBridge, ipcRenderer } = require('electron')

console.log('autoscript-preload')

// Define a new API object for communication with main process
contextBridge.exposeInMainWorld('api',
  {
    // Function to send messages to the main process
    send: (channel, data) => {
      ipcRenderer.send(channel, data);
    },
    // Function to receive messages from the main process
    receive: (channel, callback) => {
      ipcRenderer.on(channel, (event, ...args) => callback(...args));
    },
    // Function to remove IPC event listeners
    remove: (channel, callback) => {
      ipcRenderer.removeListener(channel, callback);
    },
  }
)




--- \lib\autoscripts\autoscript-server.js ---
const express = require('express');
const vm = require('vm');

// Create an Express app
const app = express();
const port = 3000;

// Define a custom evaluation function
const customEval = (cmd, context, callback) => {
  const output = [];
  // Redirect console output to the array
  const originalConsoleLog = console.log;
  console.log = (...args) => {
    output.push(args.join(' '));
  };

  try {
    // Evaluate the command using vm.runInNewContext with the provided context
    const result = vm.runInNewContext(cmd, context);
    // Restore original console.log function
    console.log = originalConsoleLog;
    // Return the result and the captured output to the callback function
    callback(null, { result, output });
  } catch (err) {
    // Restore original console.log function
    console.log = originalConsoleLog;
    // Handle any errors that occur during evaluation
    callback(err, null);
  }
};

// Configure the initial context with any variables or functions
const context = {
  console: console, // Expose the global console object
  require: require, // Expose the global require function
  greet: () => {
    return 'Hello, world!';
  }
};

// Parse JSON request bodies
app.use(express.json());

// Define a route to handle incoming POST requests
app.post('/eval', (req, res) => {
  const cmds = req.body.cmds; // Get the list of commands from the request body

  // Call the custom evaluation function for each command
  const results = cmds.map(cmd => {
    return new Promise((resolve, reject) => {
      customEval(cmd, context, (err, result) => {
        if (err) {
          // Handle any errors that occur during evaluation
          reject(err);
        } else {
          // Resolve with the result
          resolve(result);
        }
      });
    });
  });

  // Wait for all promises to resolve, and send the results as the response
  Promise.all(results)
    .then(results => {
      res.send(results);
    })
    .catch(err => {
      // Handle any errors that occur during evaluation
      res.status(500).send('Error: ' + err.message);
    });
});

// Start the Express server
app.listen(port, () => {
  console.log(`Server is running on port ${port}`);
});


--- \lib\autoscripts\autoscript.html ---
<!DOCTYPE html>
<html>

<head>
  <meta charset="UTF-8">
  <title>Electron Jupyter-like UI</title>
  <link rel="stylesheet" href="../node_modules/monaco-editor/min/vs/editor/editor.main.css">
  <style>
    #container {
      display: flex;
      flex-wrap: wrap;
      gap: 10px;
      flex-direction: column;
    }

    .step-container {
      flex: 1;
      display: flex;
      flex-direction: column;
      box-shadow: 0px 0px 4px 1px #c5dcf6;
      margin: 0.25rem;
    }


    .editor-container,
    .step-text-container {
      display: flex;
      flex-direction: row;
      width: 100%;
    }

    .step-text-container {
      background: #dee6ef;
    }



    .editor-container>div {
      flex: 1;
      min-width: 200px;
    }

    .editor,
    .step-text-human-language {
      flex: 1;
      min-height: 2em;
    }

    .step-text-human-language {
      margin: 0.5rem;
      display: flex;
      align-items: center;
    }

    .step-text-human-language-actions {
      display: flex;
      align-content: center;
      justify-content: center;
      align-items: center;
    }
    .step-text-human-language-actions span.material-icons {
      max-width: 25px;
      cursor: pointer;
    }




    .run-button,
    .step-text-number {
      display: flex;
      align-items: center;
      justify-content: center;
      cursor: pointer;

      border: none;
      border-radius: 1rem;
      width: 1.5rem;
      outline: none;
      margin: 0.5rem;

    }

    .step-text-number {
      font-weight: bold;
    }

    .run-button {
      background-color: #599fea;
      color: #191818;
      height: 1.5rem;
    }

    .step-response-machine-language {
      background: #383b3d;
      color: whitesmoke;
      padding: 0.25rem 1rem;
      box-shadow: inset 1px 3px 6px 0px #000000;
    }
    .hide{
      display: none;
    }
  </style>
  <link rel="stylesheet" href="../node_modules/material-design-icons/iconfont/material-icons.css"/>
</head>

<body>
  <div id="container"></div>
  <script src="../node_modules/monaco-editor/min/vs/loader.js"></script>
  <script>
    let scriptRef = { contextId: null }
    // Load Monaco Editor and its dependencies
    require.config({ paths: { 'vs': '../node_modules/monaco-editor/min/vs' } });
    window.MonacoEnvironment = { getWorkerUrl: () => './editor.worker.bundle.js' };

    function getActionsHTML(actions){
      return actions.map(action=>`
      <span class="material-icons" data-action data-action-type="${action.actionType}">
        ${action.icon}
        </span>
      `).join('')
    }

    function createScriptContainer(container, step, i) {
      step.stepContainer = document.createElement('div');
      step.stepContainer.id = 'step' + i;
      step.stepContainer.className = 'step-container';
      container.appendChild(step.stepContainer);

      var stepTextContainer = document.createElement('div');
      stepTextContainer.id = 'step-text' + i;
      stepTextContainer.className = 'step-text-container';
      step.stepContainer.appendChild(stepTextContainer);

      var stepTextNumberContainer = document.createElement('div');
      stepTextNumberContainer.id = 'step-text' + i;
      stepTextNumberContainer.className = 'step-text-number';
      stepTextNumberContainer.innerText = i
      stepTextContainer.appendChild(stepTextNumberContainer);

      var stepTextHumanLanguageContainer = document.createElement('div');
      stepTextHumanLanguageContainer.id = 'step-text' + i;
      stepTextHumanLanguageContainer.className = 'step-text-human-language';
      stepTextHumanLanguageContainer.innerText = step.humanLanguage
      stepTextHumanLanguageContainer.setAttribute( "contenteditable","true")
      stepTextContainer.appendChild(stepTextHumanLanguageContainer);

      var stepTextHumanLanguaActions = document.createElement('div');
      stepTextHumanLanguaActions.className = 'step-text-human-language-actions';
      stepTextHumanLanguaActions.innerHTML = getActionsHTML([{icon:'code',actionType:"show-script"},{icon:'keyboard_double_arrow_right',actionType:'run'}])
      stepTextContainer.appendChild(stepTextHumanLanguaActions);


      var editorContainer = document.createElement('div');
      editorContainer.id = 'editor' + i;
      editorContainer.className = 'editor-container hide';
      step.stepContainer.appendChild(editorContainer);

      step.editor = monaco.editor.create(document.createElement('div'), {
        value: step.machineLanguage,
        language: "javascript",
        scrollBeyondLastLine: false,
        wordWrap: 'on',
        lineNumbers: 'off', // Hide line numbers
        wrappingStrategy: 'advanced',
        automaticLayout: true,
        minimap: {
          enabled: false
        },
        fontSize: "12px",
        overviewRulerLanes: 0
      });

      const editorDom = step.editor.getDomNode()

      editorContainer.appendChild(editorDom);


      let ignoreEvent = false;
      const updateHeight = () => {
        console.log(editorDom.getBoundingClientRect())
        const width = editorDom.getBoundingClientRect().width
        const contentHeight = Math.min(1000, step.editor.getContentHeight());
        editorDom.style.width = `${width}px`;
        editorDom.style.height = `${contentHeight}px`;
        try {
          ignoreEvent = true;
          step.editor.layout({ width, height: contentHeight });
        } finally {
          ignoreEvent = false;
        }
      };
      step.editor.onDidContentSizeChange(updateHeight);
      // Automatically format the content upon loading
      setTimeout(() => {

        step.editor.getAction('editor.action.formatDocument').run();
      }, 1000)

      let handleHLAction = (action)=>{
        switch(action.dataset.actionType){
          case 'show-script':
            step.stepContainer.querySelector('.editor-container').classList.toggle('hide');
            step.editor.layout()
            break;
          case 'run':
            runMachineLanguage(step);
            break;
        }
      };
      for(let action of stepTextContainer.querySelectorAll('.step-text-human-language-actions [data-action]')){
        action.addEventListener('click', () => { handleHLAction(action) })
      }
    }

    function createScriptStepResponseContainer(step) {
      if (!step.responseContainer) {
        step.responseContainer = document.createElement('div');
        step.responseContainer.className = 'step-response-container';
        step.stepContainer.appendChild(step.responseContainer);
      }
      let humanLanguageResponseContainer = step.responseContainer.querySelector('.step-response-human-language')
      if (!humanLanguageResponseContainer) {
        humanLanguageResponseContainer = document.createElement('div');
        humanLanguageResponseContainer.className = 'step-response-human-language';
        step.responseContainer.appendChild(humanLanguageResponseContainer);
      }

      humanLanguageResponseContainer.innerHTML = step.humanLanguageResponse || ""

      let machineLanguageResponseContainer = step.responseContainer.querySelector('.step-response-machine-language')
      if (!machineLanguageResponseContainer) {
        machineLanguageResponseContainer = document.createElement('div');
        machineLanguageResponseContainer.className = 'step-response-machine-language';
        step.responseContainer.appendChild(machineLanguageResponseContainer);
      }

      machineLanguageResponseContainer.innerHTML = (step.machineLanguageResponse || []).join('<br/>')


    }

    async function runMachineLanguage(step) {
      const machineLanguage = step.editor.getValue(); // Get input from textarea

      // Send commands to preload script for evaluation
      window.api.send('AUTO_SCRIPT_REQUEST', {
        contextId: scriptRef.contextId,
        steps: [{
          machineLanguage,
          index: step.index
        }]
      });
    }

    require(['vs/editor/editor.main'], function () {
      // Monaco Editor is now available globally as `monaco`
      var container = document.getElementById('container');

      window.api.send('LOAD_SCRIPT_REQUEST', { contextId: scriptRef.contextId });
      window.api.receive('LOAD_SCRIPT_RESPONSE', (script) => {
        scriptRef = script
        console.log(script.steps);
        script.steps.forEach((step, index) => {
          step.index = index
          createScriptContainer(container, step, index + 1)
        })
      })

    });
  </script>
  <!-- Include the editor.worker.bundle.js file -->
  <script src="../node_modules/monaco-editor/min/vs/editor/editor.worker.bundle.js"></script>
  <script>
    // Handle IPC message from preload script with evaluation results
    window.api.receive('AUTO_SCRIPT_RESPONSE', (response) => {

      console.log(scriptRef, response)
      scriptRef.contextId = response.contextId

      for (var stepIndex in response.results) {
        scriptRef.steps[stepIndex].humanLanguageResponse = response.results[stepIndex].humanLanguageResponse
        scriptRef.steps[stepIndex].machineLanguageResponse = response.results[stepIndex].machineLanguageResponse

        createScriptStepResponseContainer(scriptRef.steps[stepIndex])
      }


    });
  </script>

</body>

</html>

--- \lib\autoscripts\autoscript.js ---
const vm = require('vm');
const { app, BrowserWindow, desktopCapturer, webFrameMain, ipcMain, BrowserView, session } = require('electron')

// Your code goes here
const pie = require('puppeteer-in-electron');
const puppeteer = require('puppeteer-core');


const path = require('path')
const util = require('util')

class AutoScript {
    scriptsIndex = {}

    async load(mainWindow, windows, options) {
        //Setup puppteer
        const browser = await pie.connect(app, puppeteer);
        for (let window of windows) {
            window.page = await pie.getPage(browser, window.view);

            //temporary fix , until the source of issue is found for automatic navigation to blank
            await window.page.goto(window.url);
            // Set screen size
            // await page.setViewport({width: 1080, height: 1024});
        }

        this.loadAutoScript(mainWindow, windows, options)
    }



    async loadAutoScript(mainWindow, windows, options) {
        let autoScriptWindow = {
            url: 'autoscript.html',
            preload: 'autoscript-preload.js',
            id: 'autoscript'
        }

        const autoScriptView = new BrowserView({
            webPreferences: {
                nodeIntegration: true,
                contextIsolation: true, // default in Electron >= 12
                preload: path.join(__dirname, autoScriptWindow.preload || 'preload.js')
            },
        })
        autoScriptView.setBounds({ x: Math.trunc(options.x), y: Math.trunc(options.y), width: Math.trunc(options.width), height: Math.trunc(options.height) })
        mainWindow.addBrowserView(autoScriptView)


        autoScriptView.webContents.loadFile('autoscripts/autoscript.html');

        autoScriptView.webContents.openDevTools()

        autoScriptView.setAutoResize({ width: true, height: true });

        ipcMain.on("AUTO_SCRIPT_REQUEST", async (event, data) => {
            console.log("AUTO_SCRIPT_REQUEST", data);
            const response = await this.handleAutoScript(windows, event, data)
            autoScriptView.webContents.send('AUTO_SCRIPT_RESPONSE', { contextId: response.contextId, error: null, results: response.groupedResults })
        })

        ipcMain.on("LOAD_SCRIPT_REQUEST", (event, data) => {
            autoScriptView.webContents.send('LOAD_SCRIPT_RESPONSE', {
                contextId: data.contextId,
                steps: [
                    {
                        humanLanguage: "Wait for Name input to appear",
                        machineLanguage: `// Wait and click on first result
                global.vistorName = 'mrl-text-input[input-id="visitor-modal-name"] input';
                await page.waitForSelector(vistorName, { timeout: 6000000 });
                `
                    },
                    {
                        humanLanguage: "Type Visitor name as manoj",
                        machineLanguage: `
                // Type into search box
                await page.type(global.vistorName, 'Manoj');
                `
                    },
                    {
                        humanLanguage: "Click 'Enter as a visitor'",
                        machineLanguage: `

                await page.click('text/Enter as a visitor')

                global.visitorInfo = await page.waitForResponse('https://app.mural.co/api/v0/visitor/');


                global.murals = await page.waitForResponse('https://app.mural.co/api/murals/');
                console.log(global.visitorInfo,global.murals)

                `
                    }

                ]
            })
        })



    }


    async handleResponse(response) {
        let requestUrl = response.url()
        if (requestUrl.slice(0, 5) == 'https') {
            if (requestUrl.indexOf('https://app.mural.co/api/v0/visitor/') >= 0) {
                console.log("URL: ", requestUrl)
                console.log("Headers: ", response.headers())
                console.log("Response: ", await response.json())


            } else if (requestUrl.indexOf('https://app.mural.co/api/murals/') >= 0) {
                console.log("URL: ", requestUrl)
                console.log("Headers: ", response.headers())
                console.log("Response: ", await response.json())


            } else {
                console.log("URL: ", requestUrl)
            }
        }
    }

    async handleRequest(request) {
        // The interception has not been handled yet. Control will pass through this guard.
        if (
            request.interceptResolutionState().action ===
            InterceptResolutionAction.AlreadyHandled
        )
            return;
        let requestUrl = request.url()
        if (requestUrl.slice(0, 5) == 'https') {
            if (requestUrl.indexOf('https://app.mural.co/api/v0/visitor/') >= 0) {
                console.log("URL: ", requestUrl)
                console.log("Headers: ", request.headers())
                // request.respond({
                //     content: 'application/json',
                //     body: JSON.stringify(mockResponseObject)
                // })
            } else {
                //console.log("URL: ", requestUrl)
            }
        }
        // The interception *MIGHT* have been handled by the first handler, we can't be sure.
        // Therefore, we must check again before calling continue() or we risk Puppeteer raising an exception.
        if (
            request.interceptResolutionState().action ===
            InterceptResolutionAction.AlreadyHandled
        )
            return;

        request.continue();
    }

    async handleLoad(page) {

        page.on('request', async interceptedRequest => {
            await handleRequest(interceptedRequest)
        })
        await page.setRequestInterception(true);
        page.on('response', async (response) => {
            //console.log('<<', response.status(), response.url())
            handleResponse(response)
        })

    }

    async handleScript(code, callback) {

    }

    customEval = (cmd, context, callback) => {
        const output = [];
        // Redirect console output to the array
        const originalConsoleLog = console.log;
        console.log = (...args) => {
            output.push(args.join(' '));
        };

        try {

            // Evaluate the command using vm.runInNewContext with the provided context
            const result = vm.runInNewContext(wrappedCode, context);

            // Restore original console.log function
            console.log = originalConsoleLog;
            // Return the result and the captured output to the callback function
            callback(null, { result, output });
        } catch (err) {
            // Restore original console.log function
            console.log = originalConsoleLog;
            // Handle any errors that occur during evaluation
            callback(err, null);
        }
    };

    async runInSandbox(code, sandbox, hasawait) {
        let output = ''
        const context = {
            setTimeout, Promise,
            console: {
                ...console,
                log: (...args) => {
                    output += args.join(' ') + '\n';
                }
            },
            // storeVars:(target)=> {
            //     return new Proxy(target, {
            //       has(target, prop) { return true; },
            //       get(target, prop) {
            //             //console.log(target, prop)
            //          return (prop in target ? target : global)[prop];
            //         }
            //     });
            // }
            // any other modules or functions you want to include in the sandbox
        };

        const script = new vm.Script(hasawait ? `
        (async()=>{
          ${code}
        })();
        `: `${code}`);

        const newContext = new vm.createContext(Object.assign(sandbox, context));

        const result = await script.runInNewContext(newContext);

        return output + '\n' + result;
    }


    async handleAutoScript(windows, event, args) {

        const steps = args.steps; // Get the list of commands from the IPC message
        let contextId = args.contextId; // Get the context ID from the IPC message

        // If context ID is not provided or is empty, generate a new unique ID
        if (!contextId || contextId.trim() === '') {
            contextId = 'context_' + Date.now();
        }

        // If the context ID doesn't exist in the contexts object, create a new context
        if (!this.scriptsIndex.hasOwnProperty(contextId)) {
            this.scriptsIndex[contextId] = {
                windows: windows.reduce((widows, window) => { return windows[window.window.id] }),
                pages: windows.map(window => window.page),
                console: console,
                global: {}
            };
            this.scriptsIndex[contextId].page = this.scriptsIndex[contextId].pages[0]
            vm.createContext(this.scriptsIndex[contextId])
        }

        // Get the context object for the provided context ID
        const context = this.scriptsIndex[contextId];

        // Call the custom evaluation function for each command in the context
        const results = []
        for (const step of steps) {
            // Wrap the code in a function
            let hasawait = step.machineLanguage.indexOf('await ') >= 0
            // Execute the code in the virtual machine
            try {
                results.push({ index: step.index, "machineLanguageResponse": await this.runInSandbox(step.machineLanguage, context, hasawait) })
            } catch (error) {
                // Restore original console.log function
                results.push({ index: step.index, "machineLanguageResponse": error })
            }


        }
        // const results = steps.map(step => {
        //   return new Promise((resolve, reject) => {
        //     customEval(step.machineLanguage, context, (err, result) => {
        //       if (err) {
        //         // Handle any errors that occur during evaluation
        //         step.machineLanguageResponse = err
        //         step.error = true
        //         reject(step);
        //       } else {
        //         // Resolve with the result and the context ID
        //         step.machineLanguageResponse = result
        //         step.success = true
        //         resolve(step);
        //       }
        //     });
        //   });
        // });

        // Wait for all promises to resolve
        const groupedResults = await Promise.all(results).then(results => {
            console.log(results)
            // Group results by command
            const groupedResults = results.reduce((steps, result) => {
                console.log(result)
                if (!steps[result.index]) {

                    steps[result.index] = {}

                }
                if (!steps[result.index].machineLanguageResponse) {
                    steps[result.index].machineLanguageResponse = [result.machineLanguageResponse];
                } else {
                    steps[result.index].machineLanguageResponse.push(result.machineLanguageResponse);
                }

                return steps;
            }, {});
            console.log(groupedResults)
            return groupedResults;
        });

        return { contextId, groupedResults };


    }
}

module.exports = AutoScript

--- \lib\childWindows\childWindow-preload.js ---


--- \lib\childWindows\childWindows.js ---

const { app, BrowserWindow, desktopCapturer, webFrameMain, ipcMain, BrowserView, session } = require('electron')

const path = require('path')
const util = require('util')
const AutoScript = require('../autoscripts/autoscript')

class ChildWindows {
    windowIndex = {};
    windowWiewMap = {};
    windows = [];
    dimensionMap;
    showScriptUI;
    autoScript;
    db;
    mainWindow;
    constructor(mainWindow, db) {
        this.db = db
        this.mainWindow = mainWindow
    }
    async getDefaultWindow() {
        return {
            id: "action:new",
            label: "New",
            url: "https://www.google.com",
            cssClass: "new",
            contentHtml: `
            <style>
            .container.new .content {
                display: flex;
                align-content: center;
                justify-content: center;
                align-items: center;
            }
            </style>
            <div>Search Google</div>
            `,
        }
    }
    async listWindows() {
        if(this.windows.length==0){
            this.windows = await this.db.query('windows')

            let defaultWindow = await this.getDefaultWindow()

            this.windows = [
                defaultWindow,
                // {
                //   url: 'https://app.mural.co/t/nexient7742/m/nexient7742/1654826404145/53e7ad2b3cc1ce692b3176440de112f1cbb95472?sender=630945a6-5eb4-4ec7-842f-b70c72f02d89',
                //   id: 'mural'
                // },
                // {
                //   url: 'https://tfs.ups.com/tfs/UpsProd/P02S/_backlogs/backlog/Claims%20Reimagined%20-%20PS/Features',
                //   id: "tfs"
                // },
                // {
                //   url: 'https://test.salesforce.com',
                //   id: "sfentry"
                // },
                // {
                //   url: 'https://upsdrive--drive1.sandbox.lightning.force.com/',
                //   id: "sfdrive1"
                // },
            ]
        }
        return this.windows
    }
    async checkWindowExists(webWindow) {
        return !!this.windows.find(window => webWindow.id == window.id)
    }
    async getWindowFromWebWindow(webWindow) {
        return this.windows.find(window => webWindow.id == window.id)
    }
    async updateWindowSize(window, webWindow) {
        window.x = webWindow.x
        window.y = webWindow.y
        window.width = webWindow.width
        window.height = webWindow.height
    }
    async getWindowView(window) {
        return this.windowWiewMap[window.id]
    }

    async loadWindows(webWindows) {
        let WIDTH = this.mainWindow.getSize()[0]
        let HEIGHT = this.mainWindow.getSize()[1]
        let windowWidth = WIDTH - 20, windowHeight = HEIGHT - 20
        let autoScriptWindowWidth = (this.showScriptUI ? 1 : 0.60) * WIDTH
        let childWindowsWidth = windowWidth - autoScriptWindowWidth;
        for (let webWindow of webWindows) {
            let existingWindow = await this.checkWindowExists(webWindow)
            if (!existingWindow) {
                this.windows.push(webWindow)
            }

            let window = await this.getWindowFromWebWindow(webWindow)
            await this.updateWindowSize(window, webWindow)

            let windowView = await this.getWindowView(window)

            if (!windowView) {
                await this.setupchildBrowserView(
                    window,
                    {
                        hsplit: 1,
                        width: childWindowsWidth,
                        height: windowHeight
                    }
                )
            } else {
                this.updateWindow(webWindow, 'resize')
            }


        }


        for (let window of this.windows) {
            this.windowIndex[window.id] = window
        }


        // new AutoScript().load(this.mainWindow, this.windows, {
        //     x: childWindowsWidth,
        //     y: 0,
        //     width: autoScriptWindowWidth,
        //     height: windowHeight,
        //     showScriptUI: this.showScriptUI
        // })


    }

    async setupChildWindows(mainWindow) {
        const childWindow = new BrowserWindow({
            parent: this.mainWindow,
            height: 600,
            width: 600,
            webPreferences: {
                nodeIntegration: false, // default in Electron >= 5
                contextIsolation: true, // default in Electron >= 12
                preload: path.join(__dirname, 'childWindow-preload.js')
            }
        })
        await childWindow.loadURL('https://uiblocks-dev-ed.lightning.force.com')
        childWindow.webContents.on(
            'did-frame-navigate',
            (event, url, httpResponseCode, httpStatusText, isMainFrame, frameProcessId, frameRoutingId) => {
                console.log('did-frame-navigate', url, httpResponseCode, httpStatusText, isMainFrame, frameProcessId, frameRoutingId)
                const frame = webFrameMain.fromId(frameProcessId, frameRoutingId)

                if (frame) {
                    const code = fs.readFileSync('inject.js', { encoding: 'utf8', flag: 'r' })
                    frame.executeJavaScript(code)
                }
            }
        )

        return [childWindow]

        //childWindow.hide()
    }

    async loadWindowView(window,view){
        try {
            if (window.url.match(/https?:\/\/(www\.)?[-a-zA-Z0-9@:%._\+~#=]{2,256}\.[a-z]{2,4}\b([-a-zA-Z0-9@:%_\+.~#?&//=]*)/)) {
                console.log('loadURL', window.url)
                await view.webContents.loadURL(window.url)

            } else {
                console.log('loadFile', window.url)
                await view.webContents.loadFile(window.url)
            }
        } catch (error) {
            console.error('load Error',window.url,error)
            await view.webContents.loadFile('lib/offline/offline.html')
        }
    }

    async setupchildBrowserView(window, options) {
        try {


            let urlbase64 = Buffer.from(window.url).toString('base64')
            const ses = session.fromPartition('persist:sfeditor')
            //const cookies = await ses.cookies.get({})
            const state = await this.db.fetchState(window.id)
            const cookies = state.cookies

            //console.log("restored cookies", cookies)

            const view = new BrowserView({
                webPreferences: {
                    contextIsolation: true, // default in Electron >= 12
                    preload: path.join(__dirname, window.preload || 'childWindow-preload.js'),
                    partition: 'persist:sfeditor'
                },
            })
            this.mainWindow.addBrowserView(view)

            await this.loadWindowView(window,view)


            view.setBounds({ x: Math.trunc(window.x), y: Math.trunc(window.y), width: Math.trunc(window.width), height: Math.trunc(window.height) })

            const zoom = Math.min(window.width / 1280, window.height / 720);//Math.min(w / 1920, h / 1080);
            //view.webContents.executeJavaScript(`document.body.style.zoom = ${zoom};    document.body.style.background: #d3d3d38f;`);
            view.webContents.setZoomFactor(zoom)
            //view.setBackgroundColor('green')
            view.webContents.openDevTools()

            view.setAutoResize({ width: true, height: true });

            // add an onchange event listener for the title
            view.webContents.on('page-title-updated', async (event, title) => {
                console.log(`Title changed to: ${title}`);
                let windowPatch;
                if (window.id == 'action:new') {
                    let windowId = view.webContents.id
                    windowPatch = { id: windowId,oldid:'action:new',label: title, cssClass: "notnew" }
                    Object.assign(window, windowPatch);
                    this.windowWiewMap[window.id] = this.windowWiewMap[window.oldid]
                    delete this.windowWiewMap[window.oldid]
                    this.windows.push(await this.getDefaultWindow())
                } else {
                    windowPatch = { label: title }
                    Object.assign(window, windowPatch);
                }
                console.log('this.windowWiewMap',Object.keys(this.windowWiewMap))
                this.mainWindow.webContents.send('UPDATE-WINDOWS', this.windows);
            });



            this.windowWiewMap[window.id] = view
        } catch (error) {
            console.error(error)
        }
    }

    createCurrentRowColAsKey(currentRow, currentCol) {
        return "[" + currentRow + "][" + currentCol + "]"
    }

    getCurrentRowCol(currentCount, hsplit) {
        let currentCol = currentCount % hsplit == 0 ? hsplit : currentCount % hsplit
        let currentRow = Math.ceil(currentCount / hsplit)
        return { currentRow, currentCol }
    }

    getDimensions(count, hsplit, width, height) {
        let dimensionMap = {}
        let vsplit = Math.ceil(count / hsplit)
        console.log('grid-shape-', { hsplit, vsplit, count, width, height })
        //let WIDTH = 500
        //let HEIGHT = 500
        let currentCount = 0
        for (let i = 1; i <= count; i++) {
            let { currentRow, currentCol } = this.getCurrentRowCol(i, hsplit)
            let w = (width / hsplit)
            let h = (height / vsplit)
            let x = w * (currentCol - 1);
            let y = h * (currentRow - 1);

            dimensionMap[this.createCurrentRowColAsKey(currentRow, currentCol)] = { currentRow, currentCol, x, y, w, h, hsplit, vsplit }
            //console.log(currentRow,currentCol,x,y,w,h)
        }
        return dimensionMap;
    }

    async handleNewWindow(){

    }

    async closeWindow(webWindow) {
        let window = await this.getWindowFromWebWindow(webWindow)
        let windowView = await this.getWindowView(window)

        if (windowView) {
            ///windowView.destroy(); TypeError: windowView.destroy is not a function
            windowView.webContents.destroy()
            delete this.windowWiewMap[webWindow.id]
        }
    }

    async updateWindow(webWindow, action) {
        let window = await this.getWindowFromWebWindow(webWindow)
        let windowView = await this.getWindowView(window)

        switch (action) {
            case "resize":
                if (window && windowView) {
                    await this.updateWindow(webWindow, window)
                    windowView.setBounds({ x: Math.trunc(webWindow.x), y: Math.trunc(webWindow.y), width: Math.trunc(webWindow.width), height: Math.trunc(webWindow.height) })
                }
                break;

            default:
                break;
        }
    }


}

module.exports = ChildWindows

--- \lib\db.js ---
const fsp = require('fs/promises');
const { app, BrowserWindow, desktopCapturer, webFrameMain, ipcMain, BrowserView, session } = require('electron')
const userDataPath = app.getPath('userData')


class DB {

    async query(fileName) {
        console.log('fetcing state..')
        let data = [];
        try {
            let datafile = await fsp.readFile(userDataPath + '/' + fileName + '.json', { encoding: 'utf8', flag: 'r' })
            data = JSON.parse(datafile);
        } catch (e) {
            console.error(e)
            data = [];
        }

        return data
    }

    async insert(fileName, row) {
        console.log('fetcing state..')
        let data = [];
        try {
            let datafile = await fsp.readFile(userDataPath + '/' + fileName + '.json', { encoding: 'utf8', flag: 'r' })
            data = JSON.parse(datafile);
        } catch (e) {
            console.error(e)
            data = [];

        } finally {
            data.push(row)
            await fsp.writeFile(userDataPath + '/' + fileName + '.json', JSON.stringify(data, null, 2))
        }

        return data
    }

    async update(fileName, row, uniqueKey) {
        console.log('fetcing state..')
        let data = [];
        try {
            let datafile = await fsp.readFile(userDataPath + '/' + fileName + '.json', { encoding: 'utf8', flag: 'r' })
            data = JSON.parse(datafile);
        } catch (e) {
            console.error(e)
            data = [];

        } finally {
            data.forEach((element, index) => {
                if (element[uniqueKey || 'id'] === item[uniqueKey || 'id']) {
                    data[index] = item;
                }
            });

            await fsp.writeFile(userDataPath + '/' + fileName + '.json', JSON.stringify(data, null, 2))
        }

        return data
    }
    async saveState(documentId, state) {
        console.log('Saving state..', documentId, state)
        let savedState = {}
        try {
            console.log("userData", app.getPath('userData'), app.getPath('sessionData'))

            let statefile = await fsp.readFile(userDataPath + '/' + documentId + 'state.json', { encoding: 'utf8', flag: 'r' })
            savedState = JSON.parse(statefile);
        } catch (e) {

            savedState = {}
        }


        savedState = {
            ...savedState,
            ...{
                ...{
                    cookies: {
                        ...savedState.cookies,
                        ...state.cookies,
                        modifiedAt: new Date().toISOString()
                    }
                }
            }
        }

        await fsp.writeFile(userDataPath + '/' + documentId + '-state.json', JSON.stringify(savedState, null, 2))
        console.log('Saved state..', savedState)


    }

    async fetchState(documentId) {
        console.log('fetcing state..')
        let savedState = {}
        try {
            let statefile = await fsp.readFile(userDataPath + '/' + documentId + '-state.json', { encoding: 'utf8', flag: 'r' })
            savedState = JSON.parse(statefile);
        } catch (e) {
            console.error(e)
            savedState = {}
        }

        return savedState



    }
}

module.exports = DB

--- \lib\offline\offline.html ---
<html>
    <head>
        <title>Offline</title>
    </head>
    <body>
        You are currenty offline
    </body>
</html>

--- \lib\offline\offline.js ---
class OfflineWindow{

}

module.exports = OfflineWindow

--- \main.js ---
// Access information about media sources that can be used to capture audio
// and video from the desktop using the navigator.mediaDevices.getUserMedia API.
//
// For more info, see:
// https://electronjs.org/docs/api/desktop-capturer

//require('hazardous');

const { app, BrowserWindow, desktopCapturer, webFrameMain, ipcMain, BrowserView, session } = require('electron')

const { Buffer } = require('node:buffer');


//https://stackoverflow.com/questions/58213258/how-to-use-puppeteer-core-with-electron

const puppeteer = require("puppeteer-core");



const path = require('path')
const util = require('util')
const fs = require('fs')
const fsp = require('fs/promises');


const DB = require('./lib/db');
const ChildWindows = require('./lib/childWindows/childWindows')


const userDataPath = app.getPath('userData')
const pie = require('puppeteer-in-electron');


const HEIGHT = 400;
const WIDTH = 1200;



async function createMainWindow() {
  const mainWindow = new BrowserWindow({
    height: HEIGHT,
    width: WIDTH,
    useContentSize: true,
    webPreferences: {
      nodeIntegration: false, // default in Electron >= 5
      contextIsolation: true, // default in Electron >= 12
      preload: path.join(__dirname, 'preload.js')
    }
  })
  mainWindow.maximize();
  //mainWindow.loadFile('index.html')





  desktopCapturer.getSources({ types: ['window', 'screen'] }).then(async sources => {

    // for (const source of sources) {
    //   console.log(source.name)
    //   if (source.name === 'FlowAboard Designs.pptx - PowerPoint') {
    //     mainWindow.webContents.send('SET_SOURCE', source.id)
    //     return
    //   }
    // }
    mainWindow.webContents.send('SET_SOURCES', sources.map(source => {
      return {
        id: source.id,
        name: source.name
      }
    }))

    mainWindow.getBrowserViews().forEach(bw => {
      bw.webContents.send('SET_SOURCES', sources.map(source => {
        return {
          id: source.id,
          name: source.name
        }
      }))
    })



  })

  return mainWindow;
}




function setupCookieSubscription(window, webContents) {
  let cookies = webContents.session.cookies;
  cookies.on('changed', async function (event, cookie, cause, removed) {
    if (cookie.session && !removed) {
      let url = util.format('%s://%s%s', (!cookie.httpOnly && cookie.secure) ? 'https' : 'http', cookie.domain, cookie.path);
      console.log('url', url, cookie);
      let cookiesQ = await webContents.session.cookies.get({})
      saveState(window.id, { cookies: cookiesQ })


      cookies.set({
        url: url,
        name: cookie.name,
        value: cookie.value,
        domain: cookie.domain,
        hostOnly: cookie.hostOnly,
        path: cookie.path,
        secure: cookie.secure,
        httpOnly: cookie.httpOnly,
        session: false,//cookie.session,
        expirationDate: new Date().setDate(new Date().getDate() + 14),
        sameSite: cookie.sameSite
      }, function (err) {
        if (err) {
          log.error('Error trying to persist cookie', err, cookie);
        }
      });
    }
  });
}


async function load() {
  console.profile('Loading Process')

  await pie.initialize(app);
  app.on('login', (event, webContents, details, authInfo, callback) => {
    event.preventDefault()

  })

  console.log('Initializing...')

  //console.log('browser',browser)
  await app.whenReady();

  console.profileEnd('Loading Process')

  console.profile('Creating Window')
  let mainWindow = await createMainWindow();
  // Load the HTML file into the main window

  await mainWindow.loadFile('index.html').catch(err=>console.error(err));
  mainWindow.webContents.openDevTools()
  console.profileEnd('Creating Window')
  let db  = new DB()
  const childWindows = new ChildWindows(mainWindow,db)
  let windows = await childWindows.listWindows()

  ipcMain.on("LOAD_WINDOWS", (event, windows) => {
    console.log("LOAD_WINDOWS", windows)
    childWindows.loadWindows(windows)
  })

  mainWindow.webContents.send('LIST_WINDOWS', windows)
  mainWindow.webContents.on('did-finish-load', (e) => {
    console.log('Web page loaded')
    mainWindow.webContents.send('LIST_WINDOWS', windows)
  })

  ipcMain.on("ADD_WINDOW", (event, window) => {
    console.log("ADD_WINDOW", window)
    windows.push(window)
    childWindows.loadWindows(windows)
  })

  ipcMain.on("UPDATE_WINDOW", (event, window, action) => {
    console.log("UPDATE_WINDOW", window)
    childWindows.updateWindow(window,action)
  })
  ipcMain.handle("CLOSE_WINDOW", (event, window, action) => {
    console.log("CLOSE_WINDOW", window)
    childWindows.closeWindow(window)
  })


}

load()







--- \package.json ---
{
  "name": "projecteditor2",
  "productName": "projecteditor2",
  "description": "My Electron application description",
  "keywords": [],
  "main": "./main.js",
  "version": "1.0.0",
  "author": "manchoud1",
  "scripts": {
    "start": "electron ."
  },
  "dependencies": {
    "electron-builder": "24.1.0",
    "express": "4.18.2",
    "hazardous": "0.3.0",
    "material-design-icons": "3.0.1",
    "material-icons": "1.13.4",
    "monaco-editor": "0.37.0-dev.20230322",
    "puppeteer-core": "19.8.0",
    "puppeteer-in-electron": "3.0.5"
  },
  "devDependencies": {
    "electron": "21.4.4"
  }
}

--- \preload.js ---
// In the preload script.
const { contextBridge, ipcRenderer } = require('electron')


console.log('Running Preload')
contextBridge.exposeInMainWorld('theNameYouWant',
  {
    send: (channel, payload) => ipcRenderer.send(channel, payload)
  }
)

contextBridge.exposeInMainWorld('webWindows', [])







ipcRenderer.on('LIST_WINDOWS', async (event, webWindows) => {
  window.webWindows = webWindows
  document.querySelector('.web-windows-container').innerHTML = ''
  await listwebWindows(window.webWindows)
})

ipcRenderer.on('UPDATE-WINDOWS', async (event, webWindows) => {
  //window.webWindows = webWindows

  await updatewebWindows(window.webWindows)
  document.querySelector('.web-windows-container').innerHTML = ''
  await listwebWindows(window.webWindows)
})



ipcRenderer.on('SET_SOURCE', async (event, sourceId) => {
  try {
    const stream = await navigator.mediaDevices.getUserMedia({
      audio: false,
      video: {
        mandatory: {
          chromeMediaSource: 'desktop',
          chromeMediaSourceId: sourceId,
          minWidth: 1280,
          maxWidth: 1280,
          minHeight: 720,
          maxHeight: 720
        }
      }
    })
    handleStream(stream)
  } catch (e) {
    handleError(e)
  }
})
console.log('Running Preload')
ipcRenderer.on('SET_SOURCES', async (event, sources) => {
  try {
    return
    sources.forEach(async source => {
      const stream = await navigator.mediaDevices.getUserMedia({
        audio: false,
        video: {
          mandatory: {
            chromeMediaSource: 'desktop',
            chromeMediaSourceId: source.id,
            minWidth: 1280,
            maxWidth: 1280,
            minHeight: 720,
            maxHeight: 720
          }
        }
      })
      let streamComponent = handleStream(stream, source)
    })
  } catch (e) {
    handleError(e)
  }
})
console.log('Running Preload')

function handleStream(stream, source) {
  const container = document.createElement('div')
  document.body.appendChild(container)
  container.data = source
  const title = document.createElement('div')
  container.appendChild(title)
  title.innerHTML = `<span>${source.name}<span>`
  title.setAttribute('class', 'title')
  const video = document.createElement('video')
  video.setAttribute('class', 'video')
  container.appendChild(video)

  video.srcObject = stream
  video.onloadedmetadata = (e) => video.play()

  container.addEventListener("click", () => {
    console.log(container.data)
  })

  return container;

}

async function createwebWindowsContainer() {
  const container = document.createElement('div')
  container.setAttribute('class', 'web-windows-container')
  document.body.appendChild(container)
  return container
}
async function listwebWindows(webWindows) {

  webWindows.forEach(webWindow => {

    let webWindowContainer = createwebWindow(webWindow)

    document.querySelector('.web-windows-container').appendChild(webWindowContainer)

  })
}

async function updatewebWindows(webWindows){
  window.webWindows.forEach((uiWebWindow)=>{
    let newWebWindow = webWindows.find(webWindow=>webWindow.oldid ===uiWebWindow.id)
    console.log('Updated window',uiWebWindow,newWebWindow)
    // Update UI window with backendWindow
    if(newWebWindow){

      console.log('container to update',uiWebWindow.container);

      Object.assign(uiWebWindow,newWebWindow)
      delete uiWebWindow.oldid;
    }

  })
}

async function createTopContainer(webWindowsTags) {
  const container = document.createElement('div')
  container.setAttribute('class', 'top-container')

  const topActionleftContainer = document.createElement('div')
  topActionleftContainer.setAttribute('class', 'top-action-left-container')
  container.appendChild(topActionleftContainer)

  const searchInputContainer = document.createElement('div')
  searchInputContainer.setAttribute('class', 'search-input-container')
  container.appendChild(searchInputContainer)

  const input = document.createElement('input')
  input.setAttribute('class', 'search-input')
  searchInputContainer.appendChild(input)

  const topActionRightContainer = document.createElement('div')
  topActionRightContainer.setAttribute('class', 'top-action-right-container')
  container.appendChild(topActionRightContainer)

  document.body.appendChild(container)

  autocomplete(input, webWindowsTags);
}


function autocomplete(inp, arr) {
  /*the autocomplete function takes two arguments,
  the text field element and an array of possible autocompleted values:*/
  var currentFocus;
  /*execute a function when someone writes in the text field:*/
  inp.addEventListener("input", function (e) {
    var a, b, i, val = this.value;
    /*close any already open lists of autocompleted values*/
    closeAllLists();
    if (!val) { return false; }
    currentFocus = -1;
    /*create a DIV element that will contain the items (values):*/
    a = document.createElement("DIV");
    a.setAttribute("id", this.id + "autocomplete-list");
    a.setAttribute("class", "autocomplete-items");
    /*append the DIV element as a child of the autocomplete container:*/
    this.parentNode.appendChild(a);
    /*for each item in the array...*/
    for (i = 0; i < arr.length; i++) {
      /*check if the item starts with the same letters as the text field value:*/
      if (arr[i].substr(0, val.length).toUpperCase() == val.toUpperCase()) {
        /*create a DIV element for each matching element:*/
        b = document.createElement("DIV");
        /*make the matching letters bold:*/
        b.innerHTML = "<strong>" + arr[i].substr(0, val.length) + "</strong>";
        b.innerHTML += arr[i].substr(val.length);
        /*insert a input field that will hold the current array item's value:*/
        b.innerHTML += "<input type='hidden' value='" + arr[i] + "'>";
        /*execute a function when someone clicks on the item value (DIV element):*/
        b.addEventListener("click", function (e) {
          /*insert the value for the autocomplete text field:*/
          inp.value = this.getElementsByTagName("input")[0].value;
          /*close the list of autocompleted values,
          (or any other open lists of autocompleted values:*/
          closeAllLists();
        });
        a.appendChild(b);
      }
    }
  });
  /*execute a function presses a key on the keyboard:*/
  inp.addEventListener("keydown", function (e) {
    var x = document.getElementById(this.id + "autocomplete-list");
    if (x) x = x.getElementsByTagName("div");
    if (e.keyCode == 40) {
      /*If the arrow DOWN key is pressed,
      increase the currentFocus variable:*/
      currentFocus++;
      /*and and make the current item more visible:*/
      addActive(x);
    } else if (e.keyCode == 38) { //up
      /*If the arrow UP key is pressed,
      decrease the currentFocus variable:*/
      currentFocus--;
      /*and and make the current item more visible:*/
      addActive(x);
    } else if (e.keyCode == 13) {
      /*If the ENTER key is pressed, prevent the form from being submitted,*/
      e.preventDefault();
      if (currentFocus > -1) {
        /*and simulate a click on the "active" item:*/
        if (x) x[currentFocus].click();
      }
    }
  });
  function addActive(x) {
    /*a function to classify an item as "active":*/
    if (!x) return false;
    /*start by removing the "active" class on all items:*/
    removeActive(x);
    if (currentFocus >= x.length) currentFocus = 0;
    if (currentFocus < 0) currentFocus = (x.length - 1);
    /*add class "autocomplete-active":*/
    x[currentFocus].classList.add("autocomplete-active");
  }
  function removeActive(x) {
    /*a function to remove the "active" class from all autocomplete items:*/
    for (var i = 0; i < x.length; i++) {
      x[i].classList.remove("autocomplete-active");
    }
  }
  function closeAllLists(elmnt) {
    /*close all autocomplete lists in the document,
    except the one passed as an argument:*/
    var x = document.getElementsByClassName("autocomplete-items");
    for (var i = 0; i < x.length; i++) {
      if (elmnt != x[i] && elmnt != inp) {
        x[i].parentNode.removeChild(x[i]);
      }
    }
  }
  /*execute a function when someone clicks in the document:*/
  document.addEventListener("click", function (e) {
    closeAllLists(e.target);
  });
}

function updateSize(webWindow, rect) {
  webWindow.x = rect.x
  webWindow.y = rect.y
  webWindow.width = rect.width
  webWindow.height = rect.height
}

function createwebWindow(webWindow) {
  const container = document.createElement('div')
  container.data = webWindow
  container.classList.add('container')
  if (webWindow.cssClass)
    container.classList.add(webWindow.cssClass)


  const title = document.createElement('div')
  container.appendChild(title)
  title.innerHTML = `<span class="material-icons">
  web_asset
  </span>
  <span class ="title-text">${webWindow.label || webWindow.id}</span>
  <span class="material-icons actions minimize">
    grid_view
  </span>
  <span class="material-icons actions maximize">
    tab
  </span>
  <span class="material-icons actions close">
    close
  </span>
  <span class="material-icons actions delete">
    delete
  </span>
  `
  title.setAttribute('class', 'title')

  const webWindowBody = document.createElement('div')
  webWindowBody.setAttribute('class', 'content')
  webWindowBody.innerHTML = webWindow.contentHtml
  container.appendChild(webWindowBody)

  const resize_ob = new ResizeObserver((entries) => {
    // since we are observing only a single element, so we access the first element in entries array
    let rect = entries[0].contentRect;
    let condainerContentDimention = entries[0].target.getBoundingClientRect()

    console.log('changed dimension ', rect, condainerContentDimention);

    updateSize(webWindow, condainerContentDimention)
    updatewebWindow(webWindow, 'resize')
  });

  resize_ob.observe(webWindowBody);

  title.querySelector('.actions.close').addEventListener('click', (event) => {
    container.classList.toggle('active')
    event.stopPropagation()
    handleClose(container.data)
  })
  title.querySelector('.actions.minimize').addEventListener('click', (event) => {
    container.classList.toggle('active')
    event.stopPropagation()
  })
  title.querySelector('.actions.maximize').addEventListener('click', (event) => {
    container.classList.toggle('active')
    event.stopPropagation()
  })
  let handleWindowClick = () => {
    if (!container.data.id) {
      activateSearch()
    } else {
      container.classList.toggle('active')

      let condainerContentDimention = container.querySelector('.content').getBoundingClientRect()

      updateSize(container.data, condainerContentDimention)

      loadwebWindows([container.data])
    }
  }
  webWindowBody.addEventListener('click', handleWindowClick)
  title.querySelector('span.title-text').addEventListener('click', handleWindowClick)

  return container;

}

function activateSearch() {

}

function searchwebWindows(tag) {
  ipcRenderer.send('SEARCH_WINDOWS', tag)
}

function loadwebWindows(webWindows) {
  ipcRenderer.send('LOAD_WINDOWS', webWindows)
}


function updatewebWindow(webWindow, action, params) {
  ipcRenderer.send('UPDATE_WINDOW', webWindow, action)
}

function handleError(e) {
  console.log(e)
}

function handleClose(webWindow) {
  ipcRenderer.invoke('CLOSE_WINDOW', webWindow).then((result) => {
    console.log(result);
  });
}




addEventListener("load", async () => {
  window.webWindows = []
  await createTopContainer(window.webWindows.map(webWindow => webWindow.id))
  await createwebWindowsContainer()
  await listwebWindows(window.webWindows)
})



--- \test.js ---
const vm = require('vm');

async function runInSandbox(code, sandbox ,hasawait ) {
    let output = ''
    const context = {
        setTimeout, Promise,
        console: {
            ...console,
            log: (...args) => {
                output += args.join(' ') + '\n';
            }
        },
        // storeVars:(target)=> {
        //     return new Proxy(target, {
        //       has(target, prop) { return true; },
        //       get(target, prop) {
        //             //console.log(target, prop)
        //          return (prop in target ? target : global)[prop];
        //         }
        //     });
        // }
        // any other modules or functions you want to include in the sandbox
    };

    const script = new vm.Script(hasawait?`
    (async function() {
        //with(storeVars(vars)) {
      ${code}
        //}
    })();
    `:`${code}`);

    const newContext = new vm.createContext(Object.assign(sandbox, context));

    const result = await script.runInNewContext(newContext);

    return output + '\n' + result;
}

(async function () {
    const sandbox = {global:{}};
    let result = ''
    // Example usage
    code1 = `

  console.log('start')
  const result = await new Promise(resolve => {

    setTimeout(() => {
      resolve('Hello, world!');
    }, 1000);
  });
  console.log('end')
  result.toUpperCase();
  5
  async function delay() {
    console.log('Before delay');
    await setTimeout(() => console.log('Inside delay'), 2000);
    console.log('After delay');
    return 'Done';
  }
  await delay()
  console.log('dd')
  console.log('Before 2nd delay');
  const result2 = await new Promise(resolve => setTimeout(() => resolve('Delayed result'), 1000));
  console.log('After 2nd delay');
  result2;

  var as = 1;   // Stored in vars
  var bs = 2;   // Stored in vars
  for (name in this) {
    console.log("this[" + name + "]=" + this[name])
  }
  global.test=100
`

    result = await runInSandbox(code1, sandbox,true);

    console.log(result)

    code12 = `
    let a=10,b=5;
console.log( a+b )
`


    result = await runInSandbox(code12, sandbox);

    console.log(result); // 'HELLO, WORLD!'

    code123 = `
console.log( a+b +10+global.test)
`


    result = await runInSandbox(code123, sandbox);

    console.log(result); // 'HELLO, WORLD!'
})();






""")
```
